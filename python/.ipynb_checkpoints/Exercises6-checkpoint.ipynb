### **Exercise1:**

Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
data = pd.read_csv('example.csv')
```

```python
data.loc[:, data.columns != 'name'] = data.loc[:, data.columns != 'name'] + 5
```

```python
data
```

```python
for cols in data.columns:
    if cols != 'name':
        data.loc[data[cols] > 30, cols] = 30
```

```python
data
```

```python
# data.loc[:, data.columns != 'name'].where(data.loc[:, data.columns != 'name']<30, 30, inplace = True) #False replace
```

```python
data_numeric = data.loc[:, data.columns != 'name'].copy()
data_numeric.where(data.loc[:, data.columns != 'name']<30, 30, inplace = True) 
data.loc[:, data.columns != 'name'] = data_numeric
data
```

### **Exercise2:**
 - Read in the `cars.csv` file as a dataframe
 - Remove the row with the index 3, which contains a quote in the Model column.
 - Add a new column called Price with the following values: [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]
 - Replace all values in the Make column that are Ford with Toyota.
 - Find the average Length of all the vehicles.
 - Find the median Price of all the vehicles.
 - Create a new dataframe that only contains the rows where the Price is greater than 7000.
 - Create a new dataframe that only contains the rows where the Make is Toyota and the Price is greater than 7000.
 - Sort the dataframe in descending order by the Price column.


```python
data = pd.read_csv('cars.csv')
data
```

```python
data = data.drop(labels = 2, axis = 0)
data.reset_index(inplace = True)
data
```

```python
data.groupby('Model')['Length'].mean()
```

```python
data.groupby('Model')['Length'].agg(['mean', 'median'])
```

```python
data['Price'] = pd.Series([8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800])
data
```

```python
# Replace all values in the Make column that are Ford with Toyota.
data.where(data['Make'] == 'Ford', 'Toyota')
data['Make'] = data['Make'].replace('Ford', 'Toyota')
data
```

```python
# Create a new dataframe that only contains the rows where the Price is greater than 7000.
data2 = data.loc[data.Price > 7000, :]
data2
```

```python
# Create a new dataframe that only contains the rows where the Make is Toyota and the Price is greater than 7000.
data3 = data.loc[(data.Price>7000) & (data.Make == 'Toyota') , : ]
data3
```

```python
#Sort the dataframe in descending order by the Price column.
data.sort_values(by = ['Price'], ascending = False)
```
